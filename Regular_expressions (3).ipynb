Pattern


```python
text = "My telephone number is 408-555-1234"
```

```python
phone = re.search(r'\d\d\d-\d\d\d-\d\d\d\d',text)
```

```python
phone.group()
```

```python
phone_pattern = re.compile(r'(\d{3})-(\d{3})-(\d{4})')
```

```python
results = re.search(phone_pattern,text)
```

```python
# entire result
results.group()
```

```python
results.group(1)
```

```python
results.group(2)
```

```python
results.group(3)
```

```python
results.group(4)
```

The Wildcard Character

```python
re.findall(r".at","The Cat in the hat sat here")
```

```python
re.findall(r".at","The Cat went splat.")
```

```python
re.findall(r"...at","The Cat went splat.")
```

```python
# one or more non-whitespace that ends with 'at'
re.findall(r"\S+at","The Cat went splat.")
```

```python
re.findall(r"\S+at","The Cat wentsplat.")
```

```python
re.findall(r"\S+at","The Cat went        splat.")
```

Starts With and Ends With

```python
# Ends with a number
re.findall(r"\d$","This ends with a number 2")
```

```python
# Starts with a number
re.findall(r"^\d","1 is the loneliest number.")
```

Exclusion

```python
# TO exclude characters, we can use the^symbol in conjunction with a set of brackets[]
# Anything inside the bracket is excluded
```

```python
phrase = "there are 3 numbers 34 inside 5 this sentence"
```

```python
re.findall(r'[^\d]',phrase)
```

```python
re.findall(r'[^\d]+',phrase)
```

```python
test_phrase = 'This is a string! But it has punctuation. How can we remove it?'
```

```python
re.findall('[^!.? ]+',test_phrase)    # space after ?
```

```python
clean = " ".join(re.findall('[^!.? ]+',test_phrase))
```

```python
clean
```

Brackets for grouping

```python
text = 'Only find the hypen-words in this sentence. But you do not know how long-ish they are'
```

```python
import re
```

```python
re.findall(r'[\w]+-[\w]+',text)
```

```python
re.findall(r'[\w]-[\w]',text)
```

```python
re.findall(r'[\w{2}]-[\w]',text)
```

```python
re.findall(r'[\w{2}]+-[\w]+',text)
```
