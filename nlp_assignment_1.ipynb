```python
#4. Use PyPDF2 to open the file Business_Proposal.pdf. Extract the text of page 2.
import PyPDF2

# Open the PDF file
with open("Business_Proposal.pdf", "rb") as file:
    reader = PyPDF2.PdfReader(file)
    
    # Extract text from page 2 (index 1 since pages are 0-based)
    page_two_text = reader.pages[1].extract_text()

print(page_two_text)
```

```python
#5. Open the file contacts.txt in append mode. Add the text of page 2 from above to contacts.txt.
# Open the file in append mode and add the text from page 2
with open("contacts.txt", "a") as file:
    file.write("\n" + page_two_text)
```

```python
#6. Using the page_two_text variable created above, extract any email addresses that were contained in the file Business_Proposal.pdf.
import re

# Use regex to find all email addresses in the text
emails = re.findall(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", page_two_text)

print(emails)
```
