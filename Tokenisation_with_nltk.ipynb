import libraries

```python
import nltk
from nltk import word_tokenize
```

nltk includes pre-trained models which integrates well

NLTK language model

```python
nltk.download('punkt_tab')  # it is pre-trained model for tokenize
```

Preform tokenize

```python
sentence = "Today is Friday. Next week, we're coming from U.S.!"
```

```python
for token in word_tokenize(sentence):
    print(token)
```

```python
sentence.split()
```

```python
sen2 = "We're here to help you! ask the question, email support@abcnet.ax or visit us at http://www.abcnet.ax!"
```

```python
for token in word_tokenize(sen2):
    print(token)
```
